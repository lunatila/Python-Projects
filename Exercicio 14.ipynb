```python
import scipy.special as sp
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def P(r, u):
    B = 1/(np.power(u, 1.5) * sp.iv(1, np.sqrt(u)))
    return B*sp.kn(0, r*np.sqrt(u))
```

```python
u = np.linspace(0.1,1)
r = 1

y = odeint(P, r, u)
```

```python
plt.plot(u,P(r,u))
plt.xlabel('tempo')
plt.ylabel('pressão')
plt.show()
```
